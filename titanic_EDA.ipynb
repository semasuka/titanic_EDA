```python
fig, axe = plt.subplots(figsize=(15,5),
                      ncols = 1,
                      nrows = 1)

values = titanic_df["Survived"].value_counts().index
values_counts = list(titanic_df["Survived"].value_counts())

bar_container = axe.barh(values, values_counts,height=0.5,color="yb")

plt.title("The survivor count",fontsize=25,fontweight="bold")
plt.ylabel("Survived?",fontsize=16)
plt.xlabel("Number of individuals",fontsize=16)

plt.grid()
plt.show()
```

```python
fig, ((ax_embarked,ax_cabin_class),(ax_sex,ax_age),(axe_fare_class_suv,axe_fare_age_suv),(axe_male_female_suv,axe_corr_heatmap)) = plt.subplots(
                                                                            nrows=4,
                                                                            ncols=2,
                                                                            constrained_layout=True,
                                                                            figsize=(15,15))
### percentage of man and women who survived
### heat map

embarked_cities = list(titanic_df["Embarked"].value_counts().index)
no_embarked_per_city = titanic_df["Embarked"].value_counts().values
ax_embarked.barh(embarked_cities,no_embarked_per_city,color='mbc')
ax_embarked.set_title("Number of embarked per city",fontsize=18,fontweight="bold")
ax_embarked.grid()
ax_embarked.set_ylabel("Cities",fontsize=13)
ax_embarked.set_xlabel("Number of individuals",fontsize=13)




titanic_df["Cabin class"][titanic_df["Cabin class"] == 1] = "First class"
titanic_df["Cabin class"][titanic_df["Cabin class"] == 2] = "Second class"
titanic_df["Cabin class"][titanic_df["Cabin class"] == 3] = "Third class"

type_of_cabin_class = list(titanic_df["Cabin class"].value_counts().index)
count_in_cabin_class = titanic_df["Cabin class"].value_counts().values
ax_cabin_class.bar(type_of_cabin_class,count_in_cabin_class,color='ryk',width=0.5)
ax_cabin_class.set_title("Number of individuals per cabine class",fontsize=18,fontweight="bold")
ax_cabin_class.set_xlabel("Cabin class",fontsize=13)
ax_cabin_class.set_xticklabels(type_of_cabin_class)
ax_cabin_class.set_xticks(np.arange(0,4,1))
ax_cabin_class.set_ylabel("Number of individuals",fontsize=13)
ax_cabin_class.grid()

male_and_female = list(titanic_df["Sex"].value_counts().index)
male_and_female_count = titanic_df["Sex"].value_counts().values
ax_sex.pie(male_and_female_count,
           explode = (0,0.1),
           labels = male_and_female,
           autopct="%1.1f%%",
           shadow=True,
           startangle=90,
           textprops={"color":"k","fontsize":20})
ax_sex.set_title("Male/Female ratio",fontsize=18,fontweight="bold")
ax_sex.legend(loc="upper left")






ax_age.hist(titanic_df["Age"],bins=20,histtype="step",color="r")
ax_age.set_title("Age distribution",fontsize=18,fontweight="bold")
ax_age.set_xlabel("Age",fontsize=13)
ax_age.set_ylabel("Number of individuals",fontsize=13)








cabin_class = list(sorted(titanic_df["Cabin class"].value_counts().index))
cabin_class_count = np.arange(len(cabin_class))
fare_mean_per_class_suv = titanic_df.groupby(["Cabin class","Survived"])["Fare"].mean()
fare_range_label = np.arange(0, 110, 10)
bar_width = 0.35
not_suv_bar = axe_fare_class_suv.bar(cabin_class_count - bar_width/2,fare_mean_per_class_suv.values[::2],width=bar_width,label="Not Survived",color="r")
suv_bar = axe_fare_class_suv.bar(cabin_class_count + bar_width/2,fare_mean_per_class_suv.values[1::2],width=bar_width,label="Survived",color="g")
axe_fare_class_suv.set_ylabel("Mean fare price",fontsize=13)
axe_fare_class_suv.set_xlabel("Cabin class", fontsize=13)
axe_fare_class_suv.set_title("Mean fare categoried by survival and cabin class",fontsize=18,fontweight="bold")
axe_fare_class_suv.set_yticks(fare_range_label)
axe_fare_class_suv.set_xticks(cabin_class_count)
axe_fare_class_suv.set_xticklabels(cabin_class)
axe_fare_class_suv.legend()
#annotation
def autolabel(bars):
    for bar in bars:
        axe_fare_class_suv.annotate(
                            "{:.2f}".format(bar.get_height()),
                            xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                            xytext = (0,3),
                            textcoords = "offset points",
                            ha = "center",
                            va = "bottom"
                            )
autolabel(not_suv_bar)
autolabel(suv_bar)




new_titanic_df = titanic_df
new_titanic_df["Cabin class"][new_titanic_df["Cabin class"] == "First class"] = 1.0
new_titanic_df["Cabin class"][new_titanic_df["Cabin class"] == "Second class"] = 2.0
new_titanic_df["Cabin class"][new_titanic_df["Cabin class"] == "Third class"] = 3.0
new_titanic_df["Cabin class"] = new_titanic_df["Cabin class"].astype(float)   

new_titanic_df["Survived"][new_titanic_df["Survived"] == "Yes"] = 1
new_titanic_df["Survived"][new_titanic_df["Survived"] == "No"] = 0

scatter = axe_fare_age_suv.scatter(titanic_df["Age"],
                    titanic_df["Fare"],
                    s=new_titanic_df["Cabin class"] * 100,
                    alpha = 0.5,
                    c=titanic_df["Survived"],
                    marker="o",
                    cmap = "RdYlGn",
                    edgecolors="k",
                     label=new_titanic_df["Survived"]
           )
# use seaborn to change the lanbel in the legend
legend1 = axe_fare_age_suv.legend(*scatter.legend_elements(num=1),
                    loc = 1,
                    title = "Survival Status",
                    )
axe_fare_age_suv.add_artist(legend1)

handles,labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = axe_fare_age_suv.legend(handles, labels, loc="upper left", title="Cabine class")

axe_fare_age_suv.add_artist(legend1)
axe_fare_age_suv.grid()

axe_fare_age_suv.set_title("Fare categorized by survival status and cabin class",fontsize=18,fontweight="bold") 
axe_fare_age_suv.set_ylabel("Fare paid",fontsize=13)
axe_fare_age_suv.set_xlabel("Age",fontsize=13)


male_female_ratio_suv = titanic_df.groupby(by=["Survived","Sex"])["Survived"].count().to_frame()
male_stats = [int(x) for x in male_female_ratio_suv[1::2].values]
female_stats = [int(x) for x in male_female_ratio_suv[::2].values]

male_stats_norm = []
female_stats_norm = []

def normalized(list1,list2):
    for i in np.arange(len(list1)):
        male_stats_norm.append(100* list1[i]/(list1[i]+list2[i]))
        female_stats_norm.append(100-(100* list1[i]/(list1[i]+list2[i])))


normalized(male_stats,female_stats)

axe_male_female_suv.bar(np.arange(len(male_stats_norm)), male_stats_norm, width=0.5, label="Female",color='tab:red')
axe_male_female_suv.bar(np.arange(len(female_stats_norm)), female_stats_norm, width=0.5, label="Male",bottom=male_stats_norm)
axe_male_female_suv.set_facecolor("#d1d1d1")
axe_male_female_suv.set_title("Ratio of Male/Female who survived",fontsize=18,fontweight="bold")
axe_male_female_suv.set_yticks(np.arange(0,110,10))
axe_male_female_suv.set_xticks(np.arange(0,2,1))
axe_male_female_suv.set_xticklabels(["Did not survived","Survived"])
axe_male_female_suv.set_xlabel("Survival status",fontsize=13)
axe_male_female_suv.set_ylabel("ratio(in %)",fontsize=13)
axe_male_female_suv.grid(axis="y")
axe_male_female_suv.legend()


new_titanic_df_2 = titanic_df
new_titanic_df_2["Cabin class"][new_titanic_df_2["Cabin class"] == "First class"] = 1
new_titanic_df_2["Cabin class"][new_titanic_df_2["Cabin class"] == "Second class"] = 2
new_titanic_df_2["Cabin class"][new_titanic_df_2["Cabin class"] == "Third class"] = 3


new_titanic_df_2["Embarked"][new_titanic_df_2["Embarked"] == "Queenstown"] = 1
new_titanic_df_2["Embarked"][new_titanic_df_2["Embarked"] == "Cherbourg"] = 2
new_titanic_df_2["Embarked"][new_titanic_df_2["Embarked"] == "Southampton"] = 3

new_titanic_df_2["Sex"][new_titanic_df_2["Sex"] == "male"] = 1
new_titanic_df_2["Sex"][new_titanic_df_2["Sex"] == "female"] = 2


#new_titanic_df_2.drop(labels="PassengerId",axis=1,inplace=True)

heart_map = axe_corr_heatmap.imshow(new_titanic_df_2.corr(),cmap = "hot")
axe_corr_heatmap.set_xticks(np.arange(len(new_titanic_df_2.columns)-4))
axe_corr_heatmap.set_yticks(np.arange(len(new_titanic_df_2.columns)-4))
axe_corr_heatmap.set_xticklabels(list(list(new_titanic_df_2.corr().columns)),fontsize=13)
axe_corr_heatmap.set_yticklabels(list(list(new_titanic_df_2.corr().columns)),fontsize=13)

plt.setp(axe_corr_heatmap.get_xticklabels(),rotation=45,ha="right",rotation_mode="anchor")

for i in range(len(list(new_titanic_df_2.corr().columns))):
    for j in range(len(list(new_titanic_df_2.corr().columns))):
        axe_corr_heatmap.text(j,i,np.round(new_titanic_df_2.corr().values[i,j],2),ha="center",va="center",color="w",fontsize=13,fontweight="bold")
axe_corr_heatmap.set_title("Correlation between columns",fontsize=18,fontweight="bold")
axe_corr_heatmap.set_xlabel("columns",fontsize=13)
axe_corr_heatmap.set_ylabel("columns",fontsize=13)
plt.show();
```
